```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 20,10
```

```python
lang = pd.read_csv('Most Popular Programming Languages from 2004 to 2020 V2.csv')
lang.head(10)
```

```python
pd.set_option('display.max_columns', None)
lang.describe()
```

```python
plt.plot(lang.Date, lang.Python)
plt.title('   %')
plt.xlabel('')
plt.ylabel('%')
```

          ,    .

```python
def year_ (date_):
    year_n = date_.split()[1]
    return year_n
```

```python
lang ['Year']=lang['Date'].apply(year_)
lang_year = lang.groupby('Year').mean()
```

```python
lang_year.head(20)
```

```python
ax = lang_year.plot(title='    %')
plt.xlabel('') #  xlabel    ,       'Rectangle' object has no property 'xlabel',           - 
plt.ylabel('')
```

     ,    .    ,       3%,                ,      .      ,   .

```python
Langs = list()
for Lang in lang_year.columns:
    if lang_year[Lang].mean() > 3:
        plt.plot(lang_year.index, lang_year[Lang])
        Langs.append(Lang)
plt.title('    %')
plt.xlabel('')
plt.ylabel('')
plt.legend(Langs, loc='upper left')
```

  ,   Visual Basic        2020    .    2004   .  Python   ,    2016    .  Java  PHP  . PHP        . 

```python
def small_lang():
    sm_langs = list()
    for i in lang_year.columns:
        if lang_year[i].mean() <3:
            sm_langs.append(i)
    return sm_langs
            
```

```python

lang_year.drop(small_lang(), axis=1).plot(kind='bar', stacked=True)
```

             -  ( 3%),   ,    (  ),    2012               Visual Basic, Java, PHP,       Python,        R, GO, Kotlin.     C\C++, Java, JavaScript, PHP, VisualBasic   2020    2004 . Python  C# .
