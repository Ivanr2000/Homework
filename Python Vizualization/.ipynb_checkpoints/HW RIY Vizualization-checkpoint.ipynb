```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 20,10
```

```python
lang = pd.read_csv('Most Popular Programming Languages from 2004 to 2020 V2.csv')
lang.head(10)
```

```python
pd.set_option('display.max_columns', None)
lang.describe()
```

```python
plt.plot(lang.Date, lang.Python)
plt.title('Популярность языка в %')
plt.xlabel('Дата')
plt.ylabel('%')
```

Построение графика по датам не читаемо в части подписей оси х, поэтому приведем к годам.

```python
def year_ (date_):
    year_n = date_.split()[1]
    return year_n
```

```python
lang ['Year']=lang['Date'].apply(year_)
lang_year = lang.groupby('Year').mean()
```

```python
lang_year.head(20)
```

```python
ax = lang_year.plot(title='Популярность языка программирования в %')
plt.xlabel('Год') #Не использую xlabel прямо в первой строке, так как у меня валится ошибка 'Rectangle' object has no property 'xlabel', при запуске примеров с лекции так же примеры не воспроизводятся из-за ошибки
plt.ylabel('Процент')
```

При построении по всем языкам сразу, график становится плохо читаемым. Тем не менее видно, что у большинства языков популярность меньше 3%, но тем не менее эта популярность сохраняется из года в год из чего можно сделать вывод, что они необходимы для узкоспециализированных задач. Построим отдельный график для популярных языков, популярность можно задавать.

```python
Langs = list()
for Lang in lang_year.columns:
    if lang_year[Lang].mean() > 3:
        plt.plot(lang_year.index, lang_year[Lang])
        Langs.append(Lang)
plt.title('Популярность языка программирования в %')
plt.xlabel('Год')
plt.ylabel('Процент')
plt.legend(Langs, loc='upper left')
```

По графику видно, что популярность Visual Basic падает из года в год и на 2020 язык вышел из популярных. При этом в 2004 язык достаточно популярен. Популярность Python идет постоянно вверх, и начиная с 2016 года идет резкий рост. Популярность Java и PHP постепенно снижается. PHP в текущий момент на уровне других популярных языков. 

```python
def small_lang():
    sm_langs = list()
    for i in lang_year.columns:
        if lang_year[i].mean() <3:
            sm_langs.append(i)
    return sm_langs
            
```

```python

lang_year.drop(small_lang(), axis=1).plot(kind='bar', stacked=True)
```

Аналогично предыдущему графику выведены только языки со средней популярностью за весь период больше какого-то значения (рассмматривалось 3%), по графику видно, что остальные языки (отсутствующие на графике), постепенно начиная с 2012 года отобрали часть доли у самых популярных языков в частности за счет снижения доли Visual Basic, Java, PHP, часть доли этих языков ушла в Python, а часть если посмотреть второй график в R, GO, Kotlin. Доля в вопулярности языков C\C++, Java, JavaScript, PHP, VisualBasic снизилась к 2020 по сравнени с 2004 годом. Python и C# выросли.

```python
sns.boxplot(data=lang_year[lang_year.columns], orient="h")
```

Если судить по графику, то текущие доли популчрности Python - являются выбросами. И не очевидно, что Python в 2020 году популярнее Java, т. е. или надо уменьшать объем анализируемых данных по годам, или данный вид графика не позволяет сделать однозначные выводы по текущей популярности. При этом позволяет увидеть наиболее популярные языки и наименнее популярные. 

```python
sns.heatmap(lang_year)
```

На тепловой карте хорошо видны наимболее популярные языки программирования и изменение их популярности из года в год. Хорошо видно снижение популярности Java и PHP и рост популярности Python. С\С++ и C# колеблются +\- то растут, то падуют в популярности. Javascript - держится на одном уровне. R потихоньку растет.
