```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', None)
horses = pd.read_csv ('horse_data.csv', names = ['surgery', 'Age', 'Hospital Number', 'rectal temperature', 'pulse', 
                                                 'respiratory rate', 'temperature of extremities', 'peripheral pulse', 
                                                 'mucous membranes', 'capillary refill time', 'pain', 'peristalsis', 
                                                 'abdominal distension', 'nasogastric tube', 'nasogastric reflux', 
                                                 'nasogastric reflux PH', 'rectal examination - feces', 'abdomen', 
                                                 'packed cell volume', 'total protein', 'abdominocentesis appearance', 
                                                 'abdomcentesis total protein', 'outcome', 'surgical lesion?', 'type of lesion',
                                                'type of lesion 1', 'type of lesion 2', 'cp_data'])
horses_na = horses.replace('?', np.NaN)
horses_na = horses_na.astype ('float')
#horses_na = horses_na.astype ('category')
horses_na.head(20)
```

```python
horses_des =horses_na.describe()
horses_des.loc["mediana"] = list (horses_na.median())
horses_des.loc["range"] = list (horses_na.max()-horses_na.min())
horses_des.loc["mode"]=horses_na.round().mode().loc[0]
```

1. surgery - Больше половины лошадей подверглись операционному вмешательству.
2. Age - 1. Большинство лошадей в выборке, взрослые лошади. 2. В данных есть ошибка, по легенде должно быть 2 значения "1 взрослые", "2 - молодые, младше 6 месяцев", при этом в данных отсутствует 2, но присутсвует 9, т. е. скорее всего при вводе все 2ки заменили на 9ки
3. Hospital Number - некоторые лошади попадали на лечение больше одного раза, видно при группировке по данной колонке
4. rectal temperature - данные температуры зафиксированы не у всех лошадей, отсутствует 5-ая часть данных, у тех у которых зафиксированы в большинстве случаев при болезни прослеживается повышенная температура, выше 38, при нормальной 37,8. У молодых лошадей, нормальная температура при болезние не наблюдается. 
5.  pulse - при нормальном пульсе 30-40, ни одна из лошадей при болезни не имела пониженный пульс. У молодых лошадей либо повышенный пульс у изначально, чего нет в описании, либо при болезни у абсолютно всех лошадей был повышенный пулься, у части лошадей пульс значительно выше. У взрослых лошадей при болезнях, у небольшого количества лошадей держится нормальный пульс, у большинства лошадей пульс повышается от стандартного в 1,5 - 2 раза.
6.  respiratory rate - при обозначенном в описании нормальном уровне 8-10, как и сказано в описании полезность данных сомнительна, так как отличия от нормального уровня доходят до 9-10 раз. При этом у молодых лошадей в стандартный диапазон не попадает ни одна лошадь. При этом при болезнях частота дыхания повышается в среднем в 2-3 раза. 
7.  temperature of extremities - у молодых лошадей температура конченостей при болезни делится примерно пополам, у половины нормальная, у половины пониженная, у взрослых лошадей в большинстве случаев температура конечностей понижена при болезни.
8.  peripheral pulse - значение периферийного пульса разделились примерно пополам, при этом половина имеет нормальный пульс, половина пониженный, т. е. имеют бедную перфузию (по описанию while reduced or absent indicate poor perfusion)
9.  mucous membranes - слизистая оболочка в большинстве случаев не имеет отклонений от нормального цвета, при этом для молодых лошадей это более ярко выражено. Для взрослых менее ярко. Достаточно большое количество взрослых лошадей имеют сепсис.
10. capillary refill time - судя по описанию возможные значения 1 и 2, тем не менее есть два значения "3", что скорее всего говорит об ошибках ввода. В набор ввели значение времени наполнения капиляров равное 3 секундам, т. е. должны были проставить "2". В целом примерно в двух третьих случаев, при болезни у лошадей нет проблем с циркуляцией крови. 
11. pain - суждение в описании "чем сильнее боль, тем больше вероятность хирургического вмешательства" соответствуют 1-му пункту, больше половины лошадей имели хирургическое вмешательство и больше половины лошадей имели боль среднюю и выше среднего.

Если исходить из постулата выше из лекции, то самое простое решение, это удаление строк с пропусками

```python
horses_na[horses_na['pain'].isnull()]
```

```python
horses_wo_na = horses_na.copy()
#surgery - заменяем на 1, так как уровень боли 4 и есть хирургичческое поражение

horses_wo_na.surgery.fillna(1, inplace = True)

#Age - пропуски отсутствуют
#rectal temperature - заполняем средним вначалн по возрастной группе и уровню боли, затем просто по возрасту. ИСходя из того,
# что симптомы для возраста и боли должны быть примерно одинаковыми. По идее надо брать и другие признаки, но надо знать как комбинировать.

horses_wo_na['rectal temperature'].fillna(horses_na.groupby(['Age','pain'])['rectal temperature'].transform('median'), inplace=True)
horses_wo_na['rectal temperature'].fillna(horses_na.groupby('Age')['rectal temperature'].transform('median'), inplace=True)

#pulse, respiratory rate, temperature of extremities, peripheral pulse, mucous membranes, capillary refill time - заполняем аналогично

horses_wo_na['pulse'].fillna(horses_na.groupby(['Age','pain']).pulse.transform('median'), inplace=True)
horses_wo_na['pulse'].fillna(horses_na.groupby('Age').pulse.transform('median'), inplace=True)
horses_wo_na['temperature of extremities'].fillna(horses_na.groupby(['Age','pain'])['temperature of extremities'].transform('median'), inplace=True)
horses_wo_na['temperature of extremities'].fillna(horses_na.groupby('Age')['temperature of extremities'].transform('median'), inplace=True)
horses_wo_na['peripheral pulse'].fillna(horses_na.groupby(['Age','pain'])['peripheral pulse'].transform('median'), inplace=True)
horses_wo_na['peripheral pulse'].fillna(horses_na.groupby('Age')['peripheral pulse'].transform('median'), inplace=True)
horses_wo_na['mucous membranes'].fillna(horses_na.groupby(['Age','pain'])['mucous membranes'].transform('median'), inplace=True)
horses_wo_na['mucous membranes'].fillna(horses_na.groupby('Age')['mucous membranes'].transform('median'), inplace=True)
horses_wo_na['capillary refill time'].fillna(horses_na.groupby(['Age','pain'])['capillary refill time'].transform('median'), inplace=True)
horses_wo_na['capillary refill time'].fillna(horses_na.groupby('Age')['capillary refill time'].transform('median'), inplace=True)

#pain - заполняем по моде высчитанной для возраста, так как не понятно, какие еще данные можно использовать для заполняения. 
#Заполняем сейчас, а не раньше, чтобы не было влияния на другие данные
horses_wo_na['pain'].fillna(horses_na.groupby('Age')['pain'].transform('median'), inplace=True)
```
