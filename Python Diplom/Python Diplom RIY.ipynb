# Проверить гипотезу, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом

```python
from scipy.stats import f_oneway
high_salary = hr_df [hr_df.salary =='high']
medium_salary = hr_df [hr_df.salary =='medium']
low_salary = hr_df [hr_df.salary =='low']
```

```python
F, p = f_oneway(high_salary.average_montly_hours, medium_salary.average_montly_hours, low_salary.average_montly_hours)
F, p
```

```python
from scipy import stats as st
result = st.ttest_ind(low_salary.average_montly_hours, medium_salary.average_montly_hours, equal_var=False)
print(result)
```

```python
result = st.ttest_ind(medium_salary.average_montly_hours, high_salary.average_montly_hours, equal_var=False)
print(result)
```

```python
result = st.ttest_ind(low_salary.average_montly_hours, high_salary.average_montly_hours, equal_var=False)
print(result)
```

При нулевой гипотезе, что в независимости от оклада, все сотрудники работают одинаково и уровни значимости в 5%, видим, что p-value принимает большие значения, не меньше 32%, то работники с высоким окладом проводят на работе больше времени.

# Разделить данные на тестовую и обучающую выборки
Построить модель LDA, предсказывающую уволился ли
сотрудник на основе имеющихся факторов (кроме department и
salary)
Оценить качество модели на тестовой выборки

```python
hr_df_lda = hr_df.drop(hr_df.columns[[6, 8, 9]], axis='columns')
left_emp = hr_df.left
hr_df_lda
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(hr_df_lda, left_emp, test_size=0.25)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
```

```python
lda.fit(X_train, y_train)
```

```python
lda.predict(X_test)
```

```python
result = pd.DataFrame([y_test.values, lda.predict(X_test)]).T
result
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test.values, lda.predict(X_test))
```

Точность модели около 76%, так как в ней по условию не учитывается один из ключевых параметров, уровень заработной платы.
