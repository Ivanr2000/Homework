```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
```

```python
df = pd.read_csv( 'adult.csv', na_values='?' ) #не заполненные значения в виде ? сразу преобразуем в NA
df.head(10)
```

```python
df.info()
```

```python
data=df[['age', 'workclass', 'education', 'occupation', 'race', 'gender', 'capital-gain', 'capital-loss', 'native-country', 'income']] #Оставляем колонки по которым будем делать предсказание и результирующий класс
data=data.dropna() #удаляем NA, чтобы убрать их влияние на предсказание
```

```python
data.head(10)
```

```python
selectedColumns = data[ [ 'age', 'workclass', 'education', 'occupation', 'race', 'gender', 'capital-gain', 'capital-loss', 'native-country' ] ]
X = pd.get_dummies( selectedColumns, columns = [ 'workclass', 'education', 'occupation', 'race', 'gender', 'native-country' ] )
X.head()
```

```python
le = LabelEncoder()
le.fit( data['income'] )
```

```python
y = pd.Series(le.transform( data['income'] ) )
y.head()
```

```python
model = LogisticRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size=0.3, random_state = 500)
```

```python
model.fit( X_train, y_train )
predictions = model.predict_proba( X_test )
```

```python
model.score(X_test, y_test)
```

```python
model.predict_proba( X_test )
```

```python
clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
svc = SVC(gamma='auto')
svc.fit(X_train, y_train)
```

```python
svc.score(X_test, y_test)
```
