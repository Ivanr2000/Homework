Задание 1.1  Как критики относятся к спортивным играм?

```python
vidgames = pd.read_csv('video_games_sales.csv')
```

```python
vidgames.head(10)
```

Нулевая гипотеза - спортивным играм критики ставят в среднем 7.0 баллов и выше

Альтернативная гипотеза - спортивным играм критики ставят в среднем менее 7.0 баллов

alpha = 0.05

```python
from scipy import stats as st
sport_gam = vidgames[vidgames.Genre=='Sports']

alpha = 0.05
print(sport_gam.Critic_Score.count())
print(sport_gam.Critic_Score.isnull().sum())
```

В связи с тем, что количество оценок по спортивным играм от критиков составляет менее 15%, то заполнение 
пробелов каким-то значениями очень сильно исказит статистику. Поэтому пустые строки отбрасываем.

```python
result = st.ttest_1samp(sport_gam.Critic_Score.dropna(), 7.0)
print(result)
```

Так как значение статистики больше нуля, то несмотря на маленькое значение p-value нулевую гипотезу не отбрасываем.

Задание 1.2 Критикам нравятся больше игры на PC или на PS4?

Нулевая гипотеза - критики относятся к PC играм так же как к играм на PS4

Альтернативная гипотеза - Критикам относятся по разному к играм на PC и PS4

alpha = 0.05

```python
PC_gam = vidgames[vidgames.Platform	=='PC']
PS4_gam = vidgames[vidgames.Platform	=='PS4']

alpha = 0.05
print(PC_gam.Critic_Score.count())
print(PC_gam.Critic_Score.isnull().sum())
print(PS4_gam.Critic_Score.count())
print(PS4_gam.Critic_Score.isnull().sum())
```

В связи с тем, что количество оценок по обеим платформам от критиков составляет менее 10%, то заполнение пробелов каким-то значениями очень сильно исказит статистику. Поэтому пустые строки отбрасываем.

```python
result = st.ttest_ind(PC_gam.Critic_Score.dropna(), PS4_gam.Critic_Score.dropna(), equal_var=False)
print(result)
```

Так как pvalue меньше уровня значимости, то нулевая гипотеза отвергается. Соответтсвенно критики относятся по разному к играм на PC и PS4

Задание 1.3 Критикам больше нравятся стрелялки или стратегии?

Нулевая гипотеза - критики относятся одинаково к стрелялкам и стратегиям

Альтернативная гипотеза - Критикам относятся по разному к стрелялкам и стратегиям

alpha = 0.05

```python
Strat_gam = vidgames[vidgames.Genre	=='Strategy']
Shooter_gam = vidgames[vidgames.Genre	=='Shooter']

alpha = 0.05
print(Strat_gam.Critic_Score.count())
print(Strat_gam.Critic_Score.isnull().sum())
print(Shooter_gam.Critic_Score.count())
print(Shooter_gam.Critic_Score.isnull().sum())
```

В связи с тем, что количество оценок по обеим жанрам от критиков составляет менее 25%, то заполнение пробелов каким-то значениями очень сильно исказит статистику. Поэтому пустые строки отбрасываем.

```python
result = st.ttest_ind(Strat_gam.Critic_Score.dropna(), Shooter_gam.Critic_Score.dropna(), equal_var=False)
print(result)
```

Так как pvalue больше уровня значимости, то нулевая гипотеза не отбрасывается, и соответственно критики относятся и к стратегиям и к стрелялкам одинаково.

```python
spam_df = pd.read_csv('spam.csv')
```

```python
spam_df.head()
```

```python
# приводим сообщения к нижнему регистру и делим сразу на два дата фрейма, один данные, второй результат (к чему относятся данные)
messages = spam_df.Message.str.lower()
category = pd.DataFrame(spam_df.Category)
```

```python
# удаляем мусорные символы
messages_wo_trash = messages.str.replace(r'[\W_]+',' ')
```

```python
# разделяем на слова
splitted_messages = messages_wo_trash.str.split()
splitted_messages
```

```python
from nltk.corpus import stopwords
stopwords_set = set(stopwords.words('english'))
```

```python
def del_sw (spl_mes):
    sw_del = []
    for text in spl_mes:
        if text not in stopwords_set:
            sw_del.append(text)
    return sw_del
```

```python
#удаляем стоп-слова
messages_without_sw = splitted_messages.apply(del_sw)
messages_without_sw
```

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
def lem_mes (mes_wo_sw):
    lem_words = []
    for word in mes_wo_sw:
        lem_words.append(wordnet_lemmatizer.lemmatize(word))
    return lem_words
```

```python
lemmatized_message = messages_without_sw.apply(lem_mes)
lemmatized_message
```
