Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
#нормальное распределение
mu = 1.5
sigma = 0.4
norm_rv_1 = sts.norm(loc=mu, scale=sigma)
norm_rv_1.rvs(size=10)
```

```python
#экспоненциальное распределение
lam = 0.1
expon_rv_1 = sts.expon(lam)
expon_rv_1.rvs(10)
```

```python
#распределение Стьюдента
n = 10
t_rv_1 = sts.t(n)
t_rv_1.rvs(10)
```

```python
x = np.linspace(-4,6,100)
pdf_n = norm_rv_1.pdf(x)
pdf_e = expon_rv_1.pdf(x)
pdf_s = t_rv_1.pdf(x)

plt.plot(x, pdf_n)
plt.plot(x, pdf_e)
plt.plot(x, pdf_s)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

Задание 3

Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. А что будет в случае 28, 28 и 28 секунд для каждого цвета?

```python
red = 54
yellow = 4
green = 28
total_light = red + yellow + green
p_r = red/total_light
p_y = yellow/total_light
p_g = green/total_light
```

```python
print ("Количество информации равно:", -(p_r * math.log2(p_r) + p_y * math.log2(p_y) + p_g * math.log2(p_g)))
```
