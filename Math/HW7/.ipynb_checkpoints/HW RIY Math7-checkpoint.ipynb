Монетку подбрасывают 15 раз, найдите вероятность выпадения простого числа орлов.

```python
k = 1
n = 15
p=0.5
x = [1, 2, 3, 5, 7, 11, 13]
```

```python
(math.factorial(n) / (math.factorial(k)*math.factorial(n-k)))*(p**k)*((1-p)**(n-k))
```

```python
def Prob_1 (simple):
    tot_prob = 0
    for k in simple:
        probability = (math.factorial(n) / (math.factorial(k)*math.factorial(n-k)))*(p**k)*((1-p)**(n-k))
        tot_prob += probability
    return (tot_prob)
```

```python
print ("Вероятность выпадения орлов в количестве раз являющимся простым числом (1, 2, 3, 5, 7, 11 или 13):", Prob_1(x))
```
