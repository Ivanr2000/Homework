```python
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
sns.set()
from hmmlearn import hmm
from hmmlearn.hmm import GaussianHMM
```

```python
text=''
for i in range (1,5):
    file_name='tom_'+str(i)+'.txt'
    with open(file_name, 'r', encoding='utf-8') as f:
        text = text+f.read().lower()
text = re.sub('[^а-я]','', text)
```

Строим матрицу переходных вероятностей. 

```python
RUSSIAN = 'абвгдеёжзийклмнопрстуфхцчшщъыьэюя'
POS = {l: i for i, l in enumerate(RUSSIAN)}
```

```python
probabilities = np.zeros((len(RUSSIAN), len(RUSSIAN)))
```

```python
for cp, cn in zip(text[:-1], text[1:]):
    probabilities[POS[cp], POS[cn]] += 1
```

```python
probabilities = (probabilities / probabilities.sum(axis=1)[:, None])
```

```python
probabilities
```

```python
df = pd.DataFrame(probabilities, index=(c for c in RUSSIAN), columns=(c for c in RUSSIAN))
```

```python
plt.figure(figsize=(14,8))
sns.heatmap(df)
plt.show()
```

Считаем вероятность встретить строку из 4-х символов, которая начинается п, а кончается на р

```python
start = POS['п']
end = POS['р']
prior = 1. / len(RUSSIAN)
for c in RUSSIAN:
    for g in RUSSIAN:
        proba = prior * probabilities[start, POS[c]] * probabilities[POS[c], end]
        if proba > 0:
            print(''.join(['п', c, g, 'р']), proba)
```

Генерируем слово, начинающееся с буквы “Б” и имеющее длину 8-мь символов

```python
s = []
start = 'б'
s.append(start)
length = 8
for i in range(length):
    index = POS[s[i]]
    next_char = np.random.choice(list(RUSSIAN), p=probabilities[index])
    s.append(next_char)

print (''.join(s))
```

Оцените ряд через HMM

```python
series = pd.read_csv("Series/mean-monthly-air-temperature-deg.csv")["Deg"]
```

```python
values = series.values.reshape(-1,1)
```

```python
for n in range(1, 14):
    model = GaussianHMM(n_components=n,
                        covariance_type="diag",
                        n_iter=1000)
    model.fit(values)
    print (n, model.score(values))
```

```python
model = GaussianHMM(n_components=8,
                        covariance_type="diag",
                        n_iter=1000)
model.fit(values)
labels = model.predict(values)
means = np.zeros_like(values)
for i in range(model.n_components):
    means[labels == i] = model.means_[i]
    
plt.figure(figsize=(12, 6))
plt.plot(values)
plt.plot(means, linewidth=3)
```
