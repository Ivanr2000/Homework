# Dogs vs Cats

https://www.kaggle.com/c/dogs-vs-cats-redux-kernels-edition

```python
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import inception_resnet_v2
import re

print(tf.__version__)
print(tf.executing_eagerly())
```

## Функции загрузки данных

```python
import os
from random import shuffle
from glob import glob

IMG_SIZE = (224, 224)  # размер входного изображения сети

train_files = glob('data/train/*.jpg')
test_files = glob('data/test/*.jpg')

# загружаем входное изображение и предобрабатываем
def load_image(path, target_size=IMG_SIZE):
    img = cv2.imread(path)[...,::-1]
    img = cv2.resize(img, target_size)
    return inception_resnet_v2.preprocess_input(img)  # предобработка для inception_resnet_v2

# функция-генератор загрузки обучающих данных с диска
def fit_generator(files, batch_size=32):
    batch_size = min(batch_size, len(files))
    while True:
        shuffle(files)
        for k in range(len(files) // batch_size):
            i = k * batch_size
            j = i + batch_size
            if j > len(files):
                j = - j % len(files)
            x = np.array([load_image(path) for path in files[i:j]])
            y = np.array([1. if os.path.basename(path).startswith('dog') else 0.
                          for path in files[i:j]])
            yield (x, y)

# функция-генератор загрузки тестовых изображений с диска
def predict_generator(files):
    while True:
        for path in files:
            yield np.array([load_image(path)])
```

## Визуализируем примеры для обучения

```python
%matplotlib inline
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(16, 8))
for i, path in enumerate(train_files[:10], 1):
    subplot = fig.add_subplot(2, 5, i)
    subplot.set_title('%s' % path.split('/')[-1])
    img = cv2.imread(path)[...,::-1]
    img = cv2.resize(img, IMG_SIZE)
    plt.imshow(img)
```

## Загружаем предобученную модель

```python
base_model_interc = base_model = inception_resnet_v2.InceptionResNetV2(weights='imagenet',
                         include_top=False,
                         input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3))
```

```python
base_model_interc.summary()
```

## Добавляем полносвязный слой

```python
# фиксируем все веса предобученной сети
for layer_interc in base_model_interc.layers:
    layer_interc.trainable = False

x_interc = base_model_interc.layers[-5].output
x_interc = tf.keras.layers.Flatten()(x_interc)
x_interc = tf.keras.layers.Dense(1,  # один выход (бинарная классификация)
                          activation='sigmoid',  # функция активации  
                          kernel_regularizer=tf.keras.regularizers.l1(1e-5))(x_interc)

model_interc = tf.keras.Model(inputs=base_model_interc.input, outputs=x_interc, name='dogs_vs_cats')
```

## Выводим архитектуру модели

```python
model_interc.summary()
```

## Компилируем модель и запускаем обучение

```python
model_interc.compile(optimizer='sgd', 
              loss='binary_crossentropy',  # функция потерь binary_crossentropy (log loss
              metrics=['accuracy'])
```

```python
val_samples_interc = 5000  # число изображений в валидационной выборке

shuffle(train_files)  # перемешиваем обучающую выборку
validation_data_interc = next(fit_generator(train_files[:val_samples_interc], val_samples_interc))
train_data = fit_generator(train_files[val_samples_interc:])  # данные читаем функцией-генератором

# запускаем процесс обучения
model_interc.fit(train_data,
          steps_per_epoch=10,  # число вызовов генератора за эпоху
          epochs=100,  # число эпох обучения
          validation_data=validation_data_interc)
```

```python
model_interc.save('cats-dogs-interc.hdf5')
```

## Предсказания на проверочной выборке

```python
test_pred_interc = model_interc.predict(
    predict_generator(test_files), steps=len(test_files))
```

```python
fig_interc = plt.figure(figsize=(16, 8))
for i, (path, score) in enumerate(zip(test_files[:10], test_pred_interc[:10]), 1):
    subplot = fig_interc.add_subplot(2, 5, i)
    subplot.set_title('%.2f %s' % (score, os.path.basename(path)))
    img = cv2.imread(path)[...,::-1]
    img = cv2.resize(img, IMG_SIZE)
    subplot.imshow(img)
```

## Готовим данные для отправки

```python
with open('submit_interc.csv', 'w') as dst:
    dst.write('id,label\n')
    for path, score in zip(test_files, test_pred_interc):
        dst.write('%s,%f\n' % (re.search('(\d+).jpg$', path).group(1), score))
```
