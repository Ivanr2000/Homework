```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
```

```python
def add_class (movies_for_class):
    if movies_for_class <=2:
        return "low"
    elif (movies_for_class > 2) and (movies_for_class <= 4):
        return "middle"
    else:
        return "high"
    return
```

```python
ratings['class'] = ratings['rating'].apply(add_class)
ratings.head(10)
```

Задание №2

```python
geo_data = {

'Центр': ['москва', 'тула', 'ярославль', 'нижний новгород'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
keywords = pd.read_csv('keywords.csv')
```

```python
def region_add (keywords):
    for region, cities in geo_data.items():
        for city in cities:
            if city in keywords:
                return region
    return 'undefined'
```

```python
keywords['region'] = keywords['keyword'].apply(region_add)
keywords [(keywords['region'] != 'undefined')]
```

Задание №3

```python
import numpy as np
year = list(np.arange(1950, 2020, 1))
```

```python
movies = pd.read_csv('movies.csv')
```

```python
def production_year (movies_year):
    for i in year:
        if str(i) in movies_year:
            return i
    return "1900"
```

```python
movies ['year'] = movies['title'].apply(production_year)
movies.tail(100)
```

```python
ratings_movie = ratings.groupby('movieId').mean()['rating']
ratings_movie.reset_index().head()
```

```python
movies_final = movies.merge(ratings_movie)
movies_final
```

```python
year_rating = movies_final.groupby('year').mean()['rating']
year_rating.reset_index().sort_values('rating', ascending=False).head(30)
```
