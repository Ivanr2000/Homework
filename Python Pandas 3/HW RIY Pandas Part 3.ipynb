Задание №1

```python
df_rat = pd.read_csv('ratings.csv')
df_rat.head()
```

```python
def max_min_time (rows_rat):
    max_time = df_rat.loc [df_rat.userId == rows_rat].max()['timestamp']
    min_time = df_rat.loc [df_rat.userId == rows_rat].min()['timestamp']
    long_time = round ((max_time-min_time)/60/60/24, 2)
    return long_time
```

```python
rat_count = df_rat.groupby('userId').count()['rating'].reset_index()
rat_more_100 = rat_count.loc [rat_count.rating > 100]
rat_more_100.head()
```

```python
rat_more_100 ['average_time_days'] = rat_more_100 ['userId'].apply(max_min_time)
```
